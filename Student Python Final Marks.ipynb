## Renaming Columns for better understanding

```python

renamed_student_python_marks_df = student_python_marks.rename(columns={'Students\'IDs': 'StudentsIDs', 'Assign 1/10       (Pandas)': 'Assign1_10', 'Assign2/10': 'Assign2_10', 'In-Class Practical / 10': 'InClassPractical_10', 'MidTerm Exam /30': 'MidTermExam_30', 'Final Exam/40': 'FinalExam_40'}, inplace=False)

print(renamed_student_python_marks_df.columns)
renamed_student_python_marks_df
```

```python
renamed_student_python_marks_df_sliced_df = renamed_student_python_marks_df[['Assign1_10', 'Assign2_10', 'InClassPractical_10', 'MidTermExam_30', 'FinalExam_40']]
renamed_student_python_marks_df_sliced_df
```

## Drop Missing Values with dropna

```python
renamed_student_python_marks_no_all_NA_df = renamed_student_python_marks_df_sliced_df.dropna(how='all', inplace=False) 
```

```python
renamed_student_python_marks_no_all_NA_df
```

```python
sns.heatmap(renamed_student_python_marks_no_all_NA_df.isnull())
```

```python
renamed_student_python_marks_no_all_NA_df
```

```python
renamed_student_python_marks_no_all_NA_df.mean()
```

```python
# check index of the null values
renamed_student_python_marks_no_all_NA_df.index[renamed_student_python_marks_no_all_NA_df.isnull().any(axis=1)]
```

```python
mean_value = renamed_student_python_marks_no_all_NA_df[['Assign1_10', 'Assign2_10', 'InClassPractical_10']].mean(axis=1)
print(renamed_student_python_marks_no_all_NA_df.columns)
#print(mean_value)

new_df =renamed_student_python_marks_no_all_NA_df[['Assign1_10', 'Assign2_10', 'InClassPractical_10']]
final_df = new_df.T.fillna(mean_value).T

final_df
```
